```python
import glob
import cv2
from random import sample
import random
```

```python
## Function for counting the number of frames in a video 
def count_frames(path):
  video = cv2.VideoCapture(path)
  total = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
  video.release()
  return total
```

```python
## Total videos for dissection class 
videos = glob.glob("/content/drive/MyDrive/Porcine/Dissection/"+ '*.mp4')
len(videos)
```

```python
## Number of frames in each video
for video in videos:
  print("Total = ",count_frames(video))
```

```python
## Path to save the extracted video images 
train_write_path = "/content/drive/MyDrive/NNDL_Project/Dataset/dissection/trainB/"
test_write_path = "/content/drive/MyDrive/NNDL_Project/Dataset/dissection/testB/"
```

```python
def sixty_percent_of_number(num):
  num = (num*60)/100
  return int(num)

def ten_percent_of_number(num):
  num = (num*10)/100
  return int(num)
```

### 2. Dissection clinical video images extraction

```python
train_img_num=0
test_img_num=0
for video_path in videos:
  total = count_frames(video_path)
  print(video_path," total = ",total)
  # if total>100:
  #   randomlist = sample(range(1,total),95)
  # else:
  #   randomlist = sample(range(1,total),95)
  train_list = sample(range(1,total),sixty_percent_of_number(total))
  train_list.sort()
  print(train_list)
  m = total-1
  ten_percent = ten_percent_of_number(total)
  print("10 = ",ten_percent)
  #test_list = random.choices([x for x in range(m) if x not in train_list],k=ten_percent)
  test_list = random.sample([x for x in range(m) if x not in train_list],ten_percent)
  test_list.sort()
  print(test_list)
  cap= cv2.VideoCapture(video_path)
  i=0
  train_index=0
  test_index=0
  while(cap.isOpened()):
      ret, frame = cap.read()
      if ret == False:
          break
      else:
        i+=1
      #print("len = ",len(train_list), "index  = ",train_index)
      if i==train_list[train_index]:
          if train_index<len(train_list)-1:
            train_index +=1
            train_img_num+=1
            cv2.imwrite(train_write_path+str(train_img_num)+'.jpg',frame)
      if i==test_list[test_index]:
          if test_index<len(test_list)-1:
            test_index +=1
            test_img_num+=1
            print("test_img_num=",test_img_num)
            cv2.imwrite(test_write_path+str(test_img_num)+'.jpg',frame)
      # if train_index+1==len(train_list) and test_index+1==len(test_list):
      #     break
  print(i,"")
  cap.release()
  cv2.destroyAllWindows()
```
